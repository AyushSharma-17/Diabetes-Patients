<b> Distribution of Outcome</b>

```python
#calculation the number of diabetic indiviuals (outcome =1) and non-diabetic indviduals(Outcome 0)
diabetic = len(data[data["Outcome"]==1])
non_diabetic = len(data[data["Outcome"]==0])

#Create a counts of diabetics and non diabetics
count=(diabetic,non_diabetic)
#pie chart
labels= ("Diabetic","Non Diabetc")
plt.pie(count,labels=labels,autopct='%1.1f%%',colors=('pink','lightgreen'))
#set the title
plt.title('Diabects vs Non diabetic Ratio')
plt.show()
```

<h2>Traning the model with the help of train test split</h2>

<h4>Train test Split<h4>

```python
X=data.drop("Outcome",axis=1)
Y= data['Outcome']
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2)
```

Train the model

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

Making Prediction

```python
prediction = model.predict(X_test)
print(prediction)
```

```python
accuracy =accuracy_score(prediction,Y_test)
print(accuracy)
```
